# 3.2 requests库的使用
request库比urllib更为强大，可以方便地获取cookies，完成登陆验证，代理操作等

````python
import requests

r = requests.get('https://www.baidu.com/')
# urllib 的 urlopen（）方法实际上是以GET方式请求网页，而requests中则用get()方法

print(type(r), '\n')
print(r.status_code, '\n')
print(type(r.text), '\n')
print(r.text, '\n')
print(r.cookies, '\n')
print(r.encoding) #
````

+ 其它的请求类型也可以直接调用相应的方法
```
r = requests.post('http://httpbin.org/post') # 上传文件
r = requests.put('http://httpbin.org/put')
r = requests.delete('http://httpbin.org/delete')
r = requests.head('http://httpbin.org/get')
r = requests.options('http://httpbin.org/get')
```

## GET 请求

````python
# 请求的链接为http://httpbin.org/get，该网站会判断如果客户端发起的是GET请求的话，它返回相应的请求信息
r = requests.get('http://httpbin.org/get')
print(r.text) # requests 会自动选择编码格式来解码来自服务器的内容
print(r.encoding) # 你可以直接通过这个关键字来手工指定编码 r.encoding = 'ISO-8859-1'
````

````python
# get请求添加参数
# 方法一
r = requests.get('http://httpbin.org/get?name=germey&age=22')
print(r.text)

# 方法二
data = {
    'name': 'germey',
    'age': 22
}
r = requests.get("http://httpbin.org/get", params=data)
# 请求的链接自动被构造成了 ……?arg1=value1&arg2=value2
print(r.text)
````

````python
r = requests.get("http://httpbin.org/get")
print(type(r), '\n')
print(type(r.text), '\n') # 把返回的网页内容解析成str格式
print(r.json(), '\n') # 网页的返回类型实际上是json格式
print(type(r.json()), '\n')
````

## 抓取网页

````python
# 请求普通的网页
import requests
import re
 
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'
}
r = requests.get("https://www.zhihu.com/explore", headers=headers)
r.encoding = 'utf-8'
print(r.text)
pattern = re.compile('explore-feed.*?question_link.*?>(.*?)</a>', re.S)
titles = re.findall(pattern, r.text)
print(titles) # 这里的正则表达式好像有点问题
````

## 抓取二进制数据

````python
r = requests.get("https://github.com/favicon.ico")
print(r.text) # 显示乱码
print('\n')
print(r.content) # 显示二进制码
with open('favicon.ico', 'wb') as f: # 保存二进制图片
    f.write(r.content)
````

## 添加headers
如果没有在headers里面指定*User-Agent*关键字，则*User-Agent*就会显示请求包的名字`python-requests/2.10.0`.大多数网页禁止这种请求
显示的error代码为`500`:
```
<html><body><h1>500 Server Error</h1>
An internal server error occured.
</body></html>
```

````python
import requests
 
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'
}
r = requests.get("https://www.zhihu.com/explore", headers=headers)
print(r.text[:1000])
````

## POST 请求
post请求将请求的数据以表单的格式提交

````python
data = {'name': 'germey', 'age': '22'}
r = requests.post("http://httpbin.org/post", data=data)
# 这里还是请求http://httpbin.org/post，该网站可以判断如果请求是POST方式，就把相关请求信息返回。
print(r.text) # form部分就是提交的数据
````

## 响应
requests的响应*Response*类包含多种方法，可以直接调用获取各种信息
+ 状态码
+ 响应头
+ cookies
+ url
+ history输出请求历史

````python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'
}
r = requests.get('http://www.jianshu.com', headers=headers)
print(type(r.status_code), r.status_code, '\n')
print(type(r.headers), r.headers, '\n')
print(type(r.cookies), r.cookies, '\n')
print(type(r.url), r.url, '\n')
print(type(r.history), r.history, '\n')
````
